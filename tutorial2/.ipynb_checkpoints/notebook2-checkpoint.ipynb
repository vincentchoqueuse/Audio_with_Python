# Time and Frequency Representations

In this tutorial, we will show how to load a sound, how to listen it and how to visualize it using time, frequency and time-frequency representation. Similarly to the first tutorial, we will use *numpy* and *matplotlib*. We will also use some function of scipy to compute the spectrum and the time-frequency respresentation.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read
```

## Load a Wavefile

The folder `src` contains two wavefile. We can load a wavefile using the read function of the module wavfile of scipy. 

<div class="alert alert-info">
    The documentation of the read function can be found <a href="https://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.io.wavfile.read.html">here</a>
</div>

```python
[Fs,x] = read('src/daftpunk.wav')
```

It is a good practice to look at the content of the two variables.

```python
print(x.shape)
print(Fs)
```

The variable `x` contains the data vector. It is a one-dimensional vector (mono channel) of size 334593 samples. We observe that the variable `Fs` contains the sampling frequency. This variable indicates that the content of the `x` vector must be played at `44100` samples by second. So basically, we have 8 seconds of audio.

## Listen the result

Let us listen the content of the audio waveform To play a sound directly in our browser, we can use the class Audio of the Ipython package. 

<div class="alert alert-info">
    The documentation of the Audio class can be found <a href="https://ipython.readthedocs.io/en/stable/api/generated/IPython.display.html#module-IPython.display">here</a>
</div>

```python
from IPython.display import Audio

Audio(x, rate=Fs) 
```

## Time Representation

In this section, we will learn how to display a sound in the time domain. To display a sound, we can use the `plot` function of the matplotlib library. I usually also create the time vector for the x-axis using the information contained in the variable `Fs`.

```python
t = np.arange(len(x))/Fs

plt.plot(t,x)
plt.xlabel("time (s)");
```

We observe that the sound contains several peeks. These peeks correspond the the kick drum. The kick drum is really important in dance music and is usually placed at the beginning of each measure.  We also see that the sound is not normalized (the amplitudes range from -30000 to 30000). To normalize the sound to [-1,1], we can apply the following transformation

```python
x_norm = x/np.max(np.abs(x))

plt.plot(t,x_norm)
plt.xlabel("time (s)");
```

## Frequency Representation

To analyse a sound, it is sometimes more convenient to look at the signal spectrum. The signal spectrum displays the  low sound components (bass) to the left and the treeble sound components to the right. To display the signal spectrum, we can use the `periodogram` function of the `scipy` library

<div class="alert alert-info">
    The documentation of the periodogram function can be found <a href="https://docs.scipy.org/doc/scipy-0.13.0/reference/generated/scipy.signal.periodogram.html">here</a>
</div>

```python
from scipy import signal

f, Pxx_den = signal.periodogram(x_norm, Fs)
plt.plot(f, Pxx_den)
plt.xlabel('frequency [Hz]')
plt.ylabel('Spectrum');
```

We observe that our sound is mostly composed of bass sounds. This is usually the case in dance music tracks. 

## Time-Frequency Representation

We have learnt how to represent a sound in the time domain or in the frequency domain. It could be interesting to obtain a mixed representation to show how the spectrum of the sound evoluate with time. This is called a time-frequency representation. The most used time-frequency representation is the spectrogram. The spectrogram can be displayed using the `spectrogram` function of numpy.

<div class="alert alert-info">
    The documentation of the spectrogram function can be found <a href="https://docs.scipy.org/doc/scipy/reference/generated/scipy.signal.spectrogram.html">here</a>
</div>


```python
f, t, Sxx = signal.spectrogram(x_norm, Fs)

plt.pcolormesh(t, f, Sxx)
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]');
```

Except the very low end, the representation seems to be really sparse. We can improve the representation by using a log scale using the `log` function of numpy.

```python
plt.pcolormesh(t, f, 20*np.log10(Sxx))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]');
```

Ok, it's better :) Now, it could be really interesting to experiment with other audio files. A signal processing engineer must be able to clearly understand these three different representations and to play with it to clearly highlight some components of interest.
